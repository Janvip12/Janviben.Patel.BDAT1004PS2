# Question3
Write a class named Marsupial

```python
class Marsupial:

    def __init__(self,x,y):
        self.list1 = []
        self.dx = x
        self.dy = y

    def put_in_pouch(self, item):
        return self.list1.append(item)

    def pouch_contents(self):
        print(self.list1) 
```

```python
M = Marsupial(0,0)
M.put_in_pouch('doll')
M.put_in_pouch('firetruck')
M.put_in_pouch('kitten')
M.pouch_contents()
```

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the 
attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the 
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and 
moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
c. overloads the __str__ operator so it behaves as shown below.

```python
class Kangaroo(Marsupial):
    def jump(self,x1,y1):
        self.dx +=x1
        self.dy +=y1
        return (self.dx,self.dy)
    def __str__(self):
        return "I am a Kangaroo located at coordinates ({0},{1})".format(self.dx,self.dy)
```

```python
k = Kangaroo(0,0)
print(k)
```

Write function collatz() that takes a positive integer x as input and prints the 
Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying 
this rule to the previous number x in the sequence

```python
def collatz(x):
    if x==1:
        print(int(x))
    elif x%2==0:
        print(int(x))
        x = x/2
        collatz(x)
    else:
        print(int(x))
        x = (3*x)+1
        collatz(x)
collatz(1)
```

# Question5
Write a recursive method binary() that takes a non-negative
integer n and prints the binary representation of integer n.

```python
def binary(n):
    if n<2:
        a = str(n)
    else:
        a = binary(n // 2) + str(n % 2)
    return a.replace("'"," ")
```

# Question7
Implement recursive function webdir() that takes as input: a URL (as a string) and
non-negative integers depth and indent. Your function should visit every web
page reachable from the starting URL web page in depth clicks or less, and print
each web page's URL. As shown below, indentation, specified by indent, should
be used to indicate the depth of a URL.

```python
from urllib.request import urlopen
from urllib.parse import urljoin
import re
from html.parser import HTMLParser

def analyze(url,depth,indent):
    print(url)           

    # obtain links in the web page
    content = urlopen(url).read().decode()
    collector = Collector(url,depth)
    collector.feed(content)
    urls = collector.getLinks() # get list of links

   

    return urls
```

```python
from urllib.parse import urljoin
from html.parser import HTMLParser
class Collector(HTMLParser):
    #collects hyperlink URLs into a list

    def __init__(self, url,depth):
        'initializes parser, the url, and a list'
        HTMLParser.__init__(self)
        self.url = url
        self.links = []
        self.depth = depth

    
    def handle_starttag(self, tag, attrs):
        'collects hyperlink URLs in their absolute format'
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http': # collect HTTP URLs
                        self.links.append(absolute)
                        
    def getLinks(self):
        'returns hyperlinks URLs in their absolute format'
        return self.links

```

```python
visited = set() # initialize visited to an empty set

def crawl2(url,depth,indent):
    '''a recursive web crawler that calls analyze()
       on every visited web page'''

    # add url to set of visited pages
    global visited     # warns the programmer 
    visited.add(url)

    # analyze() returns a list of hyperlink URLs in web page url 
    links = analyze(url,depth,indent)

    # recursively continue crawl from every link in links 
    for link in links:
        print(" ",end = " ")
        # follow link only if not visited
        if link not in visited:
            try:
                crawl2(link,depth,indent)
            except:
                pass
crawl2('http://reed.cs.depaul.edu/lperkovic/one.html',3,1) 
```
